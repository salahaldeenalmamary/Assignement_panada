```python
import pandas as pd

df = pd.read_csv('churn_data_before_processing.csv', index_col= 0) 
df.head(10)
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep='first')]
```

We can now remove the customerid column

```python

df.drop_duplicates(subset= ["CustomerId"], inplace=True)
print(df.duplicated().any())
print(df.shape)
```

```python
# Convert categorical variables to appropriate data types
df['Geography'] = pd.Categorical(df['Geography'])
df['Gender'] = pd.Categorical(df['Gender'])
```

```python
# Convert 'Exited' to binary (assuming 'Exited' is binary)
df['Exited'] = pd.to_numeric(df['Exited'], errors='coerce')
```

Checking for NaN values

```python
df.isnull().any()
```

```python
df.isna().sum() # counting NaN for each column
```

```python
# Drop rows with missing 'Surname'
df.dropna(subset=['Surname'], inplace=True)
```

```python
# Fill missing values with the mean for numeric columns
df['CreditScore'].fillna(df['CreditScore'].mean(), inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Tenure'].fillna(df['Tenure'].mean(), inplace=True)
df['Balance'].fillna(df['Balance'].mean(), inplace=True)
df['NumOfProducts'].fillna(df['NumOfProducts'].mean(), inplace=True)
df['HasCrCard'].fillna(df['HasCrCard'].mean(), inplace=True)
df['IsActiveMember'].fillna(df['IsActiveMember'].mean(), inplace=True)
df['EstimatedSalary'].fillna(df['EstimatedSalary'].mean(), inplace=True)
```

```python
# Fill missing values with the mode for categorical columns
df['Geography'].fillna(df['Geography'].mode()[0], inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

```python
# Fill missing values with the rounded mean for 'NumOfProducts'
df['NumOfProducts'].fillna(round(df['NumOfProducts'].mean()), inplace=True)
```

```python
df.isna().sum()
```

```python
# Save the cleaned DataFrame to a new CSV file
df.to_csv('churn_data_after_processing.csv', index=False)
```
